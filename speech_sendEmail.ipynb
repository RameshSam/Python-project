```python
import cv2
import os

# casePath = os.path.dirname("C:\Users\SAM\Desktop\ICore_project\application\haarcascade_frontalface_default.xml")
casepath = "C:/Users/SAM/Desktop/ICore_project/application/haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(casepath)

Cap = cv2.vdeoCapture("C:/Users/SAM/Desktop/ICore_project/1.mp4")

while True:
    success , frame = Cap.read()
    frame = cv2.resize(frame , (480,640))
    gray = cv2.cvtColor(frame ,cv2.COLOR_BGR2GRAY)
    
    faces = faceCascade.detectMultiScale(
        gray, 
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30,30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frame , (x,y) , (x+w , y+h) ,(0,255,0), 2)
    cv2.imshow("Interface",frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
    Cap.release()
    cv2.destroyAllWindows()
    cv2.CASCADE_SCALE_IMAGE
```
